```python
# Importações
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans, BisectingKMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris, load_wine
from kneed import KneeLocator
from sklearn.decomposition import PCA
```

#### Utilizamos a base de dados Iris e Wine

```python
# Carregando conjunto de dados
data_iris = load_iris()
X_iris = data_iris['data']
y_iris = data_iris['target']
feature_names_iris = data_iris['feature_names']

data_wine = load_wine()
X_wine = data_wine['data']
y_wine = data_wine['target']
feature_names_wine = data_wine['feature_names']
```

#### Visualizamos os formatos das bases de dados

```python
# Exibindo formato da base Iris
df_iris = pd.DataFrame(X_iris, columns=feature_names_iris)
df_iris.shape
```

```python
# Exibindo formato da base Wine
df_wine = pd.DataFrame(X_wine, columns=feature_names_wine)
df_wine.shape
```

#### Visualizamos os cabeçalhos das bases de dados

```python
# Cabeçalho da base Iris
df_iris.head()
```

```python
# Cabeçalho da base Wine
df_wine.head()
```

#### Verificando dados faltantes

```python
# Dados faltantes da base Iris
df_iris.isnull().sum()
```

```python
# Preenchendo dados faltantes com 0's
df_iris.fillna(0, inplace=True)
df_wine.fillna(0, inplace=True)
```

#### Normalizando dados

```python
scaler = StandardScaler()
```

```python
# Normalizando base Iris
X_iris = scaler.fit_transform(df_iris)
```

```python
# Normalizando base Wine
X_wine = scaler.fit_transform(df_wine)
```

#### Plotando bases de dados

```python
# Função de plotar
def plot_raw_data(df, dataset_name):
    plt.figure(figsize=(10, 6))
    plt.scatter(df.iloc[:, 0], df.iloc[:, 1], alpha=0.6)
    plt.xlabel(df.columns[0])
    plt.ylabel(df.columns[1])
    plt.title(f"Distribuição Original dos Dados - {dataset_name}")
    plt.show()
```

```python
# Plotando base Iris
plot_raw_data(df_iris, "Iris")
```

```python
# Plotando base Wine
plot_raw_data(df_wine, "Wine")
```

#### Aplicando PCA nas bases de dados devido ao número de dimensões

```python
# Aplicar PCA
pca = PCA(n_components=3)
```

```python
# Aplicando PCA na base Iris
X_iris_pca = pca.fit_transform(X_iris)
```

```python
# Aplicando PCA na base Wine
X_wine_pca = pca.fit_transform(X_wine)
```

#### Encontrando número ideal de Clusters para cada base de dados

```python
# Função de encontrar número ideal de clusters
def find_optimal_k(X):
    distortions = []
    K_range = range(2, 10)
    for k in K_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(X)
        distortions.append(kmeans.inertia_)
    
    kn = KneeLocator(K_range, distortions, curve="convex", direction="decreasing")
    return kn.elbow
```

```python
# Encontrando número ideal para base iris
k_iris = find_optimal_k(X_iris_pca)
k_iris
```

```python
# Encontrando número ideal para base wine
k_wine = find_optimal_k(X_wine_pca)
k_wine
```

#### Aplicando clustering

```python
# Função de aplicar clustering
def apply_clustering(X, k, dataset_name, feature_names):
    print(f"\n===== {dataset_name} =====")
    
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters_kmeans = kmeans.fit_predict(X)

    bisect_kmeans = BisectingKMeans(n_clusters=k, random_state=42, n_init=10)
    clusters_bisecting = bisect_kmeans.fit_predict(X)

    print(f"K-Means Silhouette Score: {silhouette_score(X, clusters_kmeans):.2f}")
    print(f"Bisecting K-Means Silhouette Score: {silhouette_score(X, clusters_bisecting):.2f}")
    print(f"K-Means Davies-Bouldin Score: {davies_bouldin_score(X, clusters_kmeans):.2f}")
    print(f"K-Means Calinski-Harabasz Score: {calinski_harabasz_score(X, clusters_kmeans):.2f}")

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    axes[0].scatter(X[:, 0], X[:, 1], c=clusters_kmeans, cmap="viridis", alpha=0.6)
    axes[0].set_title(f"K-Means Clusters - {dataset_name}")
    axes[0].set_xlabel(feature_names[0])
    axes[0].set_ylabel(feature_names[1])

    axes[1].scatter(X[:, 0], X[:, 1], c=clusters_bisecting, cmap="viridis", alpha=0.6)
    axes[1].set_title(f"Bisecting K-Means Clusters - {dataset_name}")
    axes[1].set_xlabel(feature_names[0])
    axes[1].set_ylabel(feature_names[1])
    
    Z = linkage(X, method='ward')
    dendrogram(Z, ax=axes[2])
    axes[2].set_title(f"Dendrograma - {dataset_name}")
    
    plt.tight_layout()
    plt.show()
```

```python
# Aplicando clustering para base iris 
apply_clustering(X_iris_pca, k_iris, "Iris", feature_names_iris)
```

```python
# Aplicando clustering para base wine
apply_clustering(X_wine_pca, k_wine, "Wine", feature_names_wine)
```

### Analisando resultados
##### A base Iris possui 3 classes no mundo real, mas ao analisarmos os dados sem considerar os rótulos reais, verificamos que os algoritmos de clusterização indicam uma quantidade diferente de grupos. 
#### Foram utilizados:
##### Método do Cotovelo (Elbow Method)
##### Silhouette Score
#### A técnica do cotovelo (Elbow Method) foi aplicada para determinar a quantidade ótima de clusters:
##### Iris: O número ideal de clusters foi 4
##### Wine: O número ideal de clusters foi 3
##### Os resultados apontaram para 4 clusters como melhor escolha matemática para a base Iris, o que sugere que a distribuição dos dados pode indicar subgrupos dentro de uma das classes originais, e para a base Wine 3 clusters como melhor escolha matemática.
#### Métricas de validação
##### Silhouette Score: Mede a separação dos clusters com base na proximidade dos pontos dentro do mesmo cluster e a distância para o cluster mais próximo. Quanto maior, melhor a separação dos clusters.
##### Davies-Bouldin Score: Mede a compacidade e separabilidade dos clusters. Quanto menor, melhor.
##### Calinski-Harabasz Score: Mede a dispersão interna e externa dos clusters, favorecendo clusters mais compactos e bem separados. Quanto maior, melhor.


### Conclusão
##### K-Means é mais rápido e direto, mas pode ser sensível à escolha de k (número de clusters).
##### Bisecting K-Means fornece uma abordagem hierárquica, sendo útil para dados que podem ter subdivisões naturais.
##### O Bisecting K-Means apresentou um desempenho ligeiramente melhor no conjunto Iris, mas pior no conjunto Wine. Isso sugere que o particionamento progressivo do Bisecting K-Means é mais adequado para conjuntos mais espaçados, mas pode falhar em bases com classes muito sobrepostas.
